### E. Судоку 4x4
Напишите программу, решающую такие судоку.

Формат ввода

1200  
0000  
0000  
0024

Формат вывода

1243  
4312  
2431  
3124

````python
mat = [[0]*4 for i in range(4)]

import sys
st = 0
for line in sys.stdin:
    line = line.rstrip()
    for i in range(len(line)):
        mat[st][i] = int(line[i])  
    st += 1

import copy
from math import sqrt
n = len(mat[0])
kwadr = int(sqrt(n))

class SudokuSolver:
    #если решение есть, то вернем его из solveHelper
    def solve(puzzle):
        solution = copy.deepcopy(puzzle)
        if SudokuSolver.solveHelper(solution):
            return solution
        return None

    def solveHelper(solution):
        minPossibleValueCountCell = None
        while True:
            minPossibleValueCountCell = None
            for rowIndex in range(n):
                for columnIndex in range(n):
                    #если есть цифра на своем месте, то ищем 0 далее
                    if solution[rowIndex][columnIndex] != 0:
                        continue
                    #нашли 0, который надо заполнить
                    #нашли возможные значения для этой клетки
                    possibleValues = SudokuSolver.findPossibleValues(rowIndex, columnIndex, solution)
                    possibleValueCount = len(possibleValues)
                    #если возможных значений нет, то у судоку решений нет
                    if possibleValueCount == 0:
                        return False
                    #если всего одно возможное решение, то пишем его в клетку, а из возможных значений убираем
                    if possibleValueCount == 1:
                        solution[rowIndex][columnIndex] = possibleValues.pop()
                    #если minPossibleValueCountCell = none или длина списка возможных значений меньше длины списка возможных клеток для значений
                    if not minPossibleValueCountCell or possibleValueCount < len(minPossibleValueCountCell[1]):
                        #положим туда клетку и возможные значения для нее
                        minPossibleValueCountCell = ((rowIndex, columnIndex), possibleValues)
            #если список пустых клеток закончился, то мы все решили
            if not minPossibleValueCountCell:
                return True
            #если в списке набралось больше 1 пары координат клеток, то пока остановим цикл и разберемся с этой клеткой
            elif 1 < len(minPossibleValueCountCell[1]):
                break
        r, c = minPossibleValueCountCell[0] #координаты пустой клетки
        for v in minPossibleValueCountCell[1]: #берем возможные значения из этой клетки
            solutionCopy = copy.deepcopy(solution)
            solutionCopy[r][c] = v #присвоим клетке одно из возможных значений
            if SudokuSolver.solveHelper(solutionCopy): #если нашлось решение от этого значения, то обновляем изначальный варик нашей цифрой
                for r in range(n):
                    for c in range(n):
                        solution[r][c] = solutionCopy[r][c]
                return True
        return False

    # поиск возможных для вставки циферок
    def findPossibleValues(rowIndex, columnIndex, puzzle):
        values = {v for v in range(1, n+1)} #список всех возможных циферок
        values -= SudokuSolver.getRowValues(rowIndex, puzzle) #убрать то, что есть в строках
        values -= SudokuSolver.getColumnValues(columnIndex, puzzle) #убрать то, что есть в столбцах
        values -= SudokuSolver.getBlockValues( rowIndex, columnIndex, puzzle ) #убрать то, что есть в квадратах
        return values #остались возможные варики для вставки

    #получим уникальные значения в строке
    #то, что уже стоит на местах и 0
    def getRowValues(rowIndex, puzzle):
        return set(puzzle[rowIndex][:])

    #получим уникальные значения в колонке, которые уже стоят на местах
    #пробег по всем строкам и указанной колонке
    def getColumnValues(columnIndex, puzzle):
        return {puzzle[i][columnIndex] for i in range(n)}

    #получим уникальные значения в квадрате, которые уже стоят на местах
    def getBlockValues(rowIndex, columnIndex, puzzle):
        blockRowStart = kwadr * (rowIndex // kwadr)
        blockColumnStart = kwadr * (columnIndex // kwadr)
        return {
            puzzle[blockRowStart + i][blockColumnStart + j]
                for i in range(kwadr)
                for j in range(kwadr)
        }

itog = SudokuSolver.solve(mat)
for i in itog:
    for j in i:
        print(j, end='')
    print()
````

### F. Сетка судоку

Обычно формирование нового судоку начинается с создания заполненного поля (ответа), после чего из ответа убирается нужное количество ячеек, формируя задачу заданной сложности.
